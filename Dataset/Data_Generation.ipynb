## GRE SCORE DATA GENERATION

```python
def fill_gre_waiver_and_score(row):
    if row['country'] == 'United States':
        gre_waiver = random.choice(['Yes', 'No', 'Optional'])
        if gre_waiver in ['Yes', 'Optional']:
            gre_score = str(random.randint(290,335))
        else:
            gre_score = "Not Required"
    else:
        gre_waiver = 'Not Applicable'
        gre_score = 'Not Applicable'
    return pd.Series([gre_waiver, gre_score], index=['GRE Waiver', 'GRE Score'])


df[['GRE Waiver', 'GRE Score']] = df.apply(fill_gre_waiver_and_score, axis=1)


```

##Fake Data Generation

```python
## Toefl score
Toefl=np.random.randint(90,118,1300)
df.drop(['Toefl'], axis=1)
df['Toefl']=Toefl

#IELTS Score
values = np.arange(start=6.0, stop=9.1, step=0.5, dtype=float)
IELTS = np.random.choice(values, size=1300)
df.drop(['IELTS'], axis=1)
df['IELTS']=IELTS

# GPA 
GPA_values = np.arange(start=2.5, stop=3.5, step=0.1, dtype=float)
GPA=np.random.choice(GPA_values, size=1300)
df.drop(['GPA'], axis=1)
df['GPA']=GPA

# Tuition fees
tuition=np.random.randint(25000,70000,1300)
df.drop(['Tuition and fees'], axis=1)
df['Tuition and fees']=tuition

# Acceptance Ratio
acceptance=np.arange(start=3, stop=50, step=0.1, dtype=float)
accp_ratio=np.random.choice(acceptance, size=1300)
df.drop(['Acceptance Ratio'], axis=1)
df['Acceptance Ratio']=accp_ratio


```

## Encoding Categorical Values 

```python
def encoding(df,arr_name,new_col,format):
  arr=df[arr_name].unique().tolist()
  dict_1 = { arr[i]:format+f'{i+1}'  for i in range(len(arr))}
  df[new_col] = df[arr_name].replace(dict_1)
```

```python
encoding(df,'university','UID',f'100')
encoding(df,'country','CID',f'CU')
encoding(df,'region','RID',f'R')
encoding(df,'size','SID',f'S')
encoding(df,'type','TID',f'Type')
encoding(df,'GRE Waiver','GRE_ID',f'GRE')
encoding(df,'research_output','RO_ID',f'RA_RO')
```
